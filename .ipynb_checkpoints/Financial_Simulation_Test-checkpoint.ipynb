```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# --- Personal Inputs --- Feature Engineering

current_investments = 500_000      # total invested assets today
monthly_contribution = 3_500       # how much you invest per month
annual_return = 0.08               # expected annual return (6%)
annual_expenses = 50_000           # yearly cost of living
safe_withdrawal_rate = 0.04        # 4%

start_date = datetime(2025, 1, 1)
```

```python
## convert to month
monthly_return = (1 + annual_return) ** (1/12) - 1
```

```python
# simulate net worth over time

months = []
balances = []

balance = current_investments
date = start_date

for month in range(1, 600):  # simulate up to 50 years
    balance = balance * (1 + monthly_return) + monthly_contribution
    
    months.append(date)
    balances.append(balance)
    
    # Check financial freedom condition
    if balance * safe_withdrawal_rate >= annual_expenses:
        break
    
    date = date + pd.DateOffset(months=1)
```

```python
#create dataframe

df = pd.DataFrame({
    "date": months,
    "investment_balance": balances
})

df["annual_income_from_assets"] = df["investment_balance"] * safe_withdrawal_rate
```

```python
#extract financial freedom date

freedom_row = df.iloc[-1]

freedom_date = freedom_row["date"]
freedom_balance = freedom_row["investment_balance"]

freedom_date, freedom_balance
```

```python
#Sanity Check If assets ≥ expenses → model is working.

print("Annual income from assets:",
      round(freedom_balance * safe_withdrawal_rate, 2))

print("Annual expenses:",
      annual_expenses)
```

```python
#Sensitivity analysis, change one variable at a time

def simulate_freedom_date(monthly_contribution):
    balance = current_investments
    date = start_date
    
    for _ in range(600):
        balance = balance * (1 + monthly_return) + monthly_contribution
        if balance * safe_withdrawal_rate >= annual_expenses:
            return date
        date += pd.DateOffset(months=1)
```

```python
#visualize

import matplotlib.pyplot as plt

plt.plot(df["date"], df["annual_income_from_assets"])
plt.axhline(y=annual_expenses)
plt.title("Path to Financial Freedom")
plt.ylabel("Annual Income from Assets")
plt.xlabel("Date")
plt.show()
```

```python
## monte carlo simulation

current_investments = 500_000
monthly_contribution = 3_500
annual_return_mean = 0.08
annual_return_std = 0.15      # market volatility
annual_expenses = 50_000
safe_withdrawal_rate = 0.04

months = 600                  # 50 years
n_simulations = 1_000
```

```python
# convert to monthly
monthly_return_mean = (1 + annual_return_mean) ** (1/12) - 1
monthly_return_std = annual_return_std / np.sqrt(12)
```

```python
#Monte Carlo Simulation Function

def simulate_path():
    balance = current_investments
    balances = []
    
    for _ in range(months):
        random_return = np.random.normal(
            monthly_return_mean,
            monthly_return_std
        )
        
        balance = balance * (1 + random_return) + monthly_contribution
        balances.append(balance)
        
        if balance * safe_withdrawal_rate >= annual_expenses:
            break
            
    return balances
```

```python
#Run simulations

all_simulations = []

for _ in range(n_simulations):
    path = simulate_path()
    all_simulations.append(path)
```

```python
#Extract financial freedom dates

freedom_months = []

for path in all_simulations:
    for i, balance in enumerate(path):
        if balance * safe_withdrawal_rate >= annual_expenses:
            freedom_months.append(i)
            break
```

```python
## months to years

freedom_years = np.array(freedom_months) / 12
```

```python
# visualize 
#Left side = optimistic futures 
# Right side = unlucky markets 
# The center mass = most likely outcome

plt.hist(freedom_years, bins=40)
plt.xlabel("Years to Financial Freedom")
plt.ylabel("Number of Simulations")
plt.title("Monte Carlo: Financial Freedom Timeline")
plt.show()
```

```python
## adding inflation and expense growth

# Inflation & expense assumptions
annual_inflation = 0.025          # 2.5%
expense_growth_real = 0.005       # lifestyle creep above inflation (0.5%)

# Convert to monthly
monthly_inflation = (1 + annual_inflation) ** (1/12) - 1
monthly_expense_growth = (1 + expense_growth_real) ** (1/12) - 1
```

```python
## expenses growing

def simulate_path_with_inflation():
    balance = current_investments
    expenses = annual_expenses
    balances = []
    expenses_over_time = []
    
    for _ in range(months):
        # Market return
        random_return = np.random.normal(
            monthly_return_mean,
            monthly_return_std
        )
        
        # Update investment balance
        balance = balance * (1 + random_return) + monthly_contribution
        
        # Update expenses
        expenses = expenses * (1 + monthly_inflation) * (1 + monthly_expense_growth)
        
        balances.append(balance)
        expenses_over_time.append(expenses)
        
        # Check freedom condition
        if balance * safe_withdrawal_rate >= expenses:
            break
            
    return balances, expenses_over_time
```

```python
## monte carlo simulation
all_balances = []
all_expenses = []
freedom_months = []

for _ in range(n_simulations):
    balances, expenses = simulate_path_with_inflation()
    all_balances.append(balances)
    all_expenses.append(expenses)
    
    for i in range(len(balances)):
        if balances[i] * safe_withdrawal_rate >= expenses[i]:
            freedom_months.append(i)
            break
```

```python
# convert to years
freedom_years = np.array(freedom_months) / 12
```

```python
# dist with inflation

plt.hist(freedom_years, bins=40)
plt.xlabel("Years to Financial Freedom")
plt.ylabel("Simulations")
plt.title("Monte Carlo Financial Freedom (With Inflation & Expense Growth)")
plt.show()
```

```python
# Expenses vs Passive Income (One Path)

sample_balances, sample_expenses = simulate_path_with_inflation()

passive_income = np.array(sample_balances) * safe_withdrawal_rate

plt.figure(figsize=(10, 6))
plt.plot(sample_expenses, label="Annual Expenses")
plt.plot(passive_income, label="Passive Income (4%)")
plt.xlabel("Months")
plt.ylabel("Dollars")
plt.title("Expenses vs Passive Income Over Time")
plt.legend()
plt.show()
```

```python
# New worth with inflation targets

plt.figure(figsize=(10, 6))

for balances in all_balances[:10]:
    plt.plot(balances, alpha=0.3)

# Inflation-adjusted target (final expense level)
final_expense = np.mean([e[-1] for e in all_expenses])
inflation_adjusted_target = final_expense / safe_withdrawal_rate

plt.axhline(y=inflation_adjusted_target)
plt.xlabel("Months")
plt.ylabel("Investment Balance")
plt.title("Monte Carlo Paths with Inflation-Adjusted Target")
plt.show()
```

```python
# income plus career growth

# Income assumptions
starting_income = 140_000
annual_income_growth = 0.03        # normal raises (3%)
promotion_probability = 0.10       # chance per year
promotion_raise = 0.15             # 15% jump
side_income_start_year = 5
side_income_amount = 15_000        # annual

savings_rate = 0.35                # % of income invested
```

```python
monthly_income_growth = (1 + annual_income_growth) ** (1/12) - 1
promotion_probability_monthly = promotion_probability / 12
```

```python
# updated simulation function with income growth

def simulate_full_path():
    balance = current_investments
    income = starting_income
    expenses = annual_expenses
    
    balances = []
    incomes = []
    expenses_over_time = []
    
    for month in range(months):
        # --- Market return ---
        random_return = np.random.normal(
            monthly_return_mean,
            monthly_return_std
        )
        
        # --- Income growth ---
        income = income * (1 + monthly_income_growth)
        
        # Promotion event
        if np.random.rand() < promotion_probability_monthly:
            income = income * (1 + promotion_raise)
        
        # Side income
        if month >= side_income_start_year * 12:
            income += side_income_amount / 12
        
        # --- Contributions ---
        monthly_contribution_dynamic = (income * savings_rate) / 12
        
        # --- Investment update ---
        balance = balance * (1 + random_return) + monthly_contribution_dynamic
        
        # --- Expense growth ---
        expenses = expenses * (1 + monthly_inflation) * (1 + monthly_expense_growth)
        
        balances.append(balance)
        incomes.append(income)
        expenses_over_time.append(expenses)
        
        # Financial freedom check
        if balance * safe_withdrawal_rate >= expenses:
            break
    
    return balances, incomes, expenses_over_time
```

```python
# run monte carlo simulations

all_balances = []
all_incomes = []
all_expenses = []
freedom_months = []

for _ in range(n_simulations):
    balances, incomes, expenses = simulate_full_path()
    
    all_balances.append(balances)
    all_incomes.append(incomes)
    all_expenses.append(expenses)
    
    for i in range(len(balances)):
        if balances[i] * safe_withdrawal_rate >= expenses[i]:
            freedom_months.append(i)
            break
```

```python
# convert to years

freedom_years = np.array(freedom_months) / 12
```

```python
# financial visual simulation

plt.hist(freedom_years, bins=40)
plt.xlabel("Years to Financial Freedom")
plt.ylabel("Simulations")
plt.title("Financial Freedom Timeline (Income Growth + Career Jumps)")
plt.show()
```

```python
# income vs expenses

sample_balances, sample_incomes, sample_expenses = simulate_full_path()

plt.figure(figsize=(10, 6))
plt.plot(sample_incomes, label="Annual Income")
plt.plot(sample_expenses, label="Annual Expenses")
plt.xlabel("Months")
plt.ylabel("Dollars")
plt.title("Income vs Expenses Over Time")
plt.legend()
plt.show()
```

```python
##Fan Chart (Percentile Bands Over Time)
#Used by central banks & pension funds
#Why it matters
#Shows uncertainty widening over time.

# Pad paths to same length
max_len = max(len(b) for b in all_balances)
paths = np.array([
    np.pad(b, (0, max_len - len(b)), constant_values=np.nan)
    for b in all_balances
])

percentiles = np.nanpercentile(paths, [10, 25, 50, 75, 90], axis=0)

plt.figure(figsize=(10,6))
plt.fill_between(range(max_len), percentiles[0], percentiles[4], alpha=0.2)
plt.fill_between(range(max_len), percentiles[1], percentiles[3], alpha=0.4)
plt.plot(percentiles[2])
plt.xlabel("Months")
plt.ylabel("Investment Balance")
plt.title("Net Worth Fan Chart")
plt.show()
```

```python
# Contributions vs Market Growth Breakdown

#Shows what’s actually doing the work

#Why it matters

#People overestimate markets and underestimate behavior.

market_growth = []
contributions = []

balance = current_investments
income = starting_income

for _ in range(120):
    prev_balance = balance
    r = np.random.normal(monthly_return_mean, monthly_return_std)
    contribution = (income * savings_rate) / 12
    
    balance = balance * (1 + r) + contribution
    
    market_growth.append(balance - prev_balance - contribution)
    contributions.append(contribution)
```

```python
plt.stackplot(range(len(market_growth)),
              contributions,
              market_growth)
plt.xlabel("Months")
plt.ylabel("Growth")
plt.title("Wealth Growth: Contributions vs Market Returns")
plt.show()
```
